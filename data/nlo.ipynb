```python
import pandas as pd

ufo_data = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv', sep=',')
ufo_data.head(30)
```

```python
ufo_data_c = ufo_data.copy()
ufo_data_d = ufo_data_c.drop(['City', 'Colors Reported', 'Shape Reported'], axis=1)
nev = ufo_data_d.loc[ufo_data_d['State'] == 'NV']
nev['Date'] = pd.to_datetime(nev['Time'])
delta = nev['Date'].diff()
delta_days = delta.dt.days
delta_days.mean()



```
