```python
import pandas as pd
import re
import numpy as np

ratings_movies = pd.read_csv('C:/IDE/data/ratings_movies.csv', sep=',')
def get_year_release(arg):
    candidates = re.findall(r'\(\d{4}\)', arg) 
    if len(candidates) > 0:
        year = candidates[0].replace('(', '')
        year = int(year.replace(')', ''))
        return year
    else:
        return 0
    
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
ratings_movies.sort_values(by=['year_release'])
ratings_movies.groupby(['year_release', 'title'])['rating'].mean().loc[1999].sort_values()
ratings_movies.groupby(['year_release', 'genres'])['rating'].mean().loc[2010].sort_values()

ratings_movies.groupby('userId')['genres'].nunique().sort_values(ascending=False)
ratings_movies.groupby(['userId'])['rating'].agg(['count', 'mean']).sort_values(['count', 'mean'], ascending=[True, False])

ratings_movies[(ratings_movies['year_release'] == 2018)].groupby('genres')['rating'].agg('count').sort_values(ascending=False) #главный вопрос был

ratings_movies['year_rating'] = pd.to_datetime(ratings_movies['date']).dt.year

ratings_movies.pivot_table(
    values='rating',
    index='year_rating',
    columns='genres',
    aggfunc=['mean'],
    fill_value=0
).loc[2018].sort_values(ascending=False) #очень важно!
```
